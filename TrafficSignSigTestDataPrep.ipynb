```python
import os
import cv2
import numpy as np
import pandas as pd
from PIL import Image
```

```python
SIGN_TEST_PATH = 'data/TrafficSigns'
FNAME1 = 'Test.csv'
SIGNAL_TEST_DAYIMGS = 'data/TrafficSignals/daySequence1/daySequence1/frames'
SIGNAL_TEST_NIGHTIMGS = 'data/TrafficSignals/nightSequence1/nightSequence1/frames'
SIGNAL_TEST_PATH1 = 'data/TrafficSignals/Annotations/daySequence1'
SIGNAL_TEST_PATH2 = 'data/TrafficSignals/Annotations/nightSequence1'
FNAME2 = 'frameAnnotationsBOX.csv'
DESTINATION_FOLDER = 'ProcessedData/Test'
DESTINATION_FOLDER_IMGS = 'ProcessedData/Test/Images'
```

```python
for img in os.listdir(DESTINATION_FOLDER_IMGS):
    os.remove(os.path.join(DESTINATION_FOLDER_IMGS, img))
```

```python
sign_annotations = pd.read_csv(os.path.join(SIGN_TEST_PATH, FNAME1))
```

```python
sign_annotations
```

```python
def changeFilename(x):
    return SIGN_TEST_PATH +'/'+ x.Path
    
sign_annotations['Path'] = sign_annotations.apply(changeFilename,axis=1)
```

```python
sign_annotations
```

```python
sign_test_df =sign_annotations.groupby(['ClassId']).apply(lambda x:x.sample(50)).reset_index(drop=True)
```

```python
sign_test_df
```

```python
for idx, row in sign_test_df.iterrows():
    image = Image.open(row['Path'])
    image.save(os.path.join(DESTINATION_FOLDER_IMGS, row['Path'].split('/')[-1]))
```

```python
def chgFname(x):
    splitted = x.Path.split('/')[-1]
    return f'{DESTINATION_FOLDER_IMGS}/{splitted}'
    
sign_test_df['Path'] = sign_test_df.apply(chgFname,axis=1)
```

```python
sign_test_df
```

```python
signal_annos_1 = pd.read_csv(os.path.join(SIGNAL_TEST_PATH1, FNAME2), sep = ";")
signal_annos_2 = pd.read_csv(os.path.join(SIGNAL_TEST_PATH2, FNAME2), sep = ";")
```

```python
signal_annos = pd.concat([signal_annos_1, signal_annos_2], ignore_index = True)
```

```python
signal_annos = signal_annos.drop(['Origin file', 'Origin frame number', 
                                              'Origin track', 'Origin track frame number'], axis = 1)
```

```python
signal_annos
```

```python
tags = {'go': 43, 'warning': 44, 'stop': 45}

def changeAnnotation(x):
    if 'go' in x['Annotation tag']:
        return tags['go']
    elif 'warning' in x['Annotation tag']:
        return tags['warning']
    elif 'stop' in x['Annotation tag']:
        return tags['stop']
    
signal_annos['Annotation tag'] = signal_annos.apply(changeAnnotation,axis=1)
```

```python
signal_test_df =signal_annos.groupby(['Annotation tag']).apply(lambda x:x.sample(50)).reset_index(drop=True)
```

```python
signal_test_df
```

```python
def chgFilename(x):
    filename = x.Filename    
    splitted = filename.split('/')
    clipName = splitted[-1].split('--')[0]
    if 'nightTest' in filename:
        return f'{SIGNAL_TEST_NIGHTIMGS}/{splitted[-1]}'
    else:
        return f'{SIGNAL_TEST_DAYIMGS}/{splitted[-1]}'

    
signal_test_df['Filename'] = signal_test_df.apply(chgFilename,axis=1)
```

```python
signal_test_df
```

```python
for idx, row in signal_test_df.iterrows():
    
    image = Image.open(row['Filename'])
    image = image.crop((row['Upper left corner X'] - 10, row['Upper left corner Y'] - 10, row['Lower right corner X'] + 10,
                        row['Lower right corner Y'] + 10))
    
    image.save(os.path.join(DESTINATION_FOLDER_IMGS, row['Filename'].split('/')[-1]))
    
    del image
```

```python
signal_annos1 = pd.DataFrame(columns = ['Width', 'Height', 'Roi.X1', 'Roi.Y1', 'Roi.X2', 'Roi.Y2', 'ClassId', 'Path'])

for idx, row in signal_test_df.iterrows():
    img = Image.open(DESTINATION_FOLDER_IMGS + '/' + row['Filename'].split('/')[-1])
    w, h = img.size
    signal_annos1 = pd.DataFrame(np.insert(signal_annos1.values, 0, values = [w, h, 10, 10, 
                                row['Lower right corner X'] - row['Upper left corner X']+10, 
                                row['Lower right corner Y'] - row['Upper left corner Y']+10,
                                row['Annotation tag'], DESTINATION_FOLDER_IMGS + '/' + row['Filename'].split('/')[-1]], 
                                axis= 0),
                            columns = ['Width', 'Height', 'Roi.X1', 'Roi.Y1', 'Roi.X2', 'Roi.Y2', 'ClassId', 'Path'])
```

```python
signal_annos1
```

```python
test_dataset = pd.concat([sign_test_df, signal_annos1], ignore_index = True)
```

```python
test_dataset
```

```python
test_dataset = test_dataset.sample(frac=1).reset_index(drop=True)
```

```python
if os.path.exists(os.path.join(DESTINATION_FOLDER, 'frameAnnotationsBOX.csv')):
    os.remove(os.path.join(DESTINATION_FOLDER, 'frameAnnotationsBOX.csv'))

test_dataset.to_csv(os.path.join(DESTINATION_FOLDER, 'frameAnnotationsBOX.csv'), index = False)
```
